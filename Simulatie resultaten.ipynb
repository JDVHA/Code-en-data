```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Model result.txt", skiprows=8)
df
```

```python
x = df["% x"]
y = df["y"]
v = df["spf.U (m/s) @ t=10"]

fig, ax = plt.subplots(figsize=(12, 4))

cnt = ax.tricontourf(x, y, v)
c = fig.colorbar(cnt)
ax.set_xlabel("x [mm]")
ax.set_ylabel("y [mm]")
c.ax.set_ylabel('$v$ [m/s]', rotation=270, labelpad=15)
fig.savefig("Figuren/Simulatie resultaat.png")
plt.show()
```
